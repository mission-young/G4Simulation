先前的数据处理中，会发现Al或者Mg的模拟谱在3MeV之后与实验谱对不上。分析可能的原因如下：
1. 调整注入深度，将z分别设置为前表面，中间，后表面。对称分布的情形下，两者应该一致。
2. 调整G4模型

调整质子只从最后一块Si发射，x,y按照第三块Si的分布，z设置为固定值。

1. z=38010 um

```c++
%jsroot on
double f1(double x)
{
    double energy[25] = {475, 721, 975, 1033, 1223, 1299, 1551, 1753, 2072, 2503, 2583, 2838, 3088, 3484, 4017, 4224, 4464, 4912, 5177, 5667, 5808, 5909, 6085, 6774, 7517};
    double br[25] = {4.73, 7.39, 0.25, 3, 0.75, 18.51, 0.81, 0.45, 0.48, 0.64, 4.89, 2.11, 1.89/4, 2.18/4, 1.078/4, 0.84/4, 3.21/4, 0.27/4, 0.29, 0.35, 0.18, 0.21, 0.41, 0.41, 0.33};
    double s1[25] = {17.91,19,20,17.77,20,21.25,20,20,20,25.5,20,20,20,36.42,20,37.21,20,20,20,20,20,20,20,20,20};
    //double s2[25] = {16.12,15.89,20,15.23,20,15.68,20,20,20,21.13,20,20,20,37.01,20,39.28,20,20,20,20,20,20,20,20,20};
   // double sig = 20;
    double c[25];
    double sum = 0;
    for (int i = 0; i < 25; i++)
    {
        c[i] = br[i] * TMath::Gaus(x, energy[i], s1[i]);
        sum += c[i];
    }
    return sum;
}
```
