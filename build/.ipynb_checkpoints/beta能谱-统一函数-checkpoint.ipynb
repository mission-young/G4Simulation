```c++
//%jsroot on
TCanvas c;
char name[32];
char tmp[128];
TF1 *bf[19];
TH1D *bh[19];
for(int i=0;i<19;i++){
    sprintf(name,"Q%02dMeV",1+i);
    sprintf(tmp,"sqrt(x*x+2*x*511)*pow(%d-x,2)*(x+511)*(x<=%d && x>=0)",(1+i)*1000,(1+i)*1000);
    bf[i]=new TF1(name,tmp,0,20000);
    bh[i]=new TH1D(name,name,200,0,20000);
    for(int j=0;j<200;j++) bh[i]->Fill(100*j,bf[i]->Eval(100*j));
}
c.Clear();
bf[18]->Draw();
for(int i=0;i<17;i++) bf[i]->Draw("same");
//c.Draw();
c.Clear();
bh[18]->Draw();
for(int i=0;i<17;i++) bh[i]->Draw("same");
//c.Draw();
for(int i=0;i<19;i++){
    bh[i]->Scale(1000000/bh[i]->Integral(0,199));
}

c.Clear();
// bh[4]->SetLineColor(1);
// bh[4]->Draw("");
// bh[6]->Draw("same");
// bh[7]->Draw("same");
// bh[8]->Draw("same");
// bh[9]->Draw("same");
// bh[10]->Draw("same");
// bh[12]->Draw("same");

bh[4]->Draw("HIST");
bh[6]->Draw("HIST same");
bh[7]->Draw("HIST same");
bh[8]->Draw("HIST same");
bh[9]->Draw("HIST same");
bh[10]->Draw("HIST same");
bh[12]->Draw("HIST same");
c.BuildLegend();
c.Draw();
```
