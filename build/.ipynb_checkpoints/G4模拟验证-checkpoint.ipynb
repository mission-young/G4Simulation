粒子能量大于2MeV时，能够从中间Si穿出。此时打到前后两块Si的总计数基本保持定值(17800),几何效率为35.6%.  

$$ \sum_{i=0}^{15}\sum_{j=0}^{15}\sum_{k=0}^{15}\sum_{l=0}^{15}\frac{3.05^2}{19^2+[3.05(k-i)]^2+[3.05(l-j)]^2} $$

```c++
void solidangle()
{
    double sum = 0;
    for (int i = 0; i < 16; i++)
        for (int j = 0; j < 16; j++)
            for (int k = 0; k < 16; k++)
                for (int l = 0; l < 16; l++)
                    sum += 3.05 * 3.05 / (19 * 19 + 3.05 * 3.05 * ((k - i) * (k - i) + (l - j) * (l - j)));
    cout << "solidangle: " << sum / 16 / 16 << '\n'
         << "eff: " << 100 * sum / 16 / 16 / 4 / 3.1415 << "%" << endl;
}
solidangle();
```

经过简单的估算，计算得出两块相距19mm的约50\*50um的DSSD的立体角为2.82, 几何效率为22%. 与模拟所得18%相近。比模拟较高的原因来源于，计算使用的面积为单块Si的面积，而立体角计算的为投影面积，因而导致计算的立体角偏大。

$$ \sum_{i=0}^{15}\sum_{j=0}^{15}\sum_{k=0}^{15}\sum_{l=0}^{15}\frac{19*3.05^2}{\{19^2+[3.05(k-i)]^2+[3.05(l-j)]^2\}^{\frac{3}{2}}} $$

```c++
void solidangle2()
{
    double sum = 0;
    for (int i = 0; i < 16; i++)
        for (int j = 0; j < 16; j++)
            for (int k = 0; k < 16; k++)
                for (int l = 0; l < 16; l++)
                    sum += 3.05 * 3.05*19 / TMath::Power((19 * 19 + 3.05 * 3.05 * ((k - i) * (k - i) + (l - j) * (l - j))),1.5);
    cout << "solidangle: " << sum / 16 / 16 << '\n'
         << "eff: " << 100 * sum / 16 / 16 / 4 / 3.1415 << "%" << endl;
}
solidangle2();
```

经过面积修正之后，探测效率降为16%，更接近模拟所得的18%。偏小的原因在于：在中间Si的16\*16个晶格(发射源)中，中间部分的立体角偏大，边缘部分的立体角偏小。在计算中是简单地对16\*16个晶格的立体角做平均，但实际上，在模拟中，这些晶格并不是平权的，中心权重大，边缘权重小。因而计算的结果会偏小。

$$ \int_{x_1=0}^{49.5}\int_{y_1=0}^{49.5}\int_{x_2=0}^{49.5}\int_{y_2=0}^{49.5}19\cdot[(x_2-x_1)^2+(y_2-y_1)^2+19^2]^{-\frac{3}{2}}\textrm{d}x_2\textrm{d}y_2 $$

```c++
void solidangle3()
{
    double sum = 0;
    double x1=0;
    double y1=0;
    double x2=0;
    double y2=0;
    double delta=1;
    for(double x1=0;x1<49.5;x1+=delta)
         for(double y1=0;y1<49.5;y1+=delta)
              for(double x2=0;x2<49.5;x2+=delta)
                   for(double y2=0;y2<49.5;y2+=delta)
                    sum+= 19*TMath::Power((x2-x1)*(x2-x1)+(y2-y1)*(y2-y1)+19*19,-1.5)*delta*delta;
   
    cout << "solidangle: " << sum/49.5/49.5<< '\n'
         << "eff: " << 100 * sum  /49.5/49.5/ 4 / 3.1415 << "%" << endl;
}
solidangle3();
```

```c++
void solidangle4()
{
    double sum = 0;
    double x1 = 0;
    double y1 = 0;
    double x2 = 0;
    double y2 = 0;
    double delta = 1;
    for (double x1 = 0; x1 < 49.5; x1 += delta)
        for (double y1 = 0; y1 < 49.5; y1 += delta)
            for (double x2 = 0; x2 < 49.5; x2 += delta)
                for (double y2 = 0; y2 < 49.5; y2 += delta)
                    sum += 38 * TMath::Power((x2 - x1) * (x2 - x1) + (y2 - y1) * (y2 - y1) + 38 * 38, -1.5) * delta * delta;

    std::cout << "solidangle: " << sum / 49.5 / 49.5 << '\n'
              << "eff: " << 100 * sum / 49.5 / 49.5 / 4 / 3.1415 << "%" << std::endl;
}
solidangle4();
```

从300umSi发射，被142um接收到的几何效率为8.22%，与20MeV下模拟获得的8.2%一致。结合142um Si发射效率，可以印证，程序获得的几何效率是有保障的。
