## G4模拟获得 beta能谱-Si探测器

程序设置细节：
1. $\beta$能谱应用[beta能谱-root内置](beta能谱-root内置.ipynb)一节中的`getabeta2()`函数生成的`beta2.root`文件作为G4模拟的e+能量输入。
2. 模拟的时候依旧是对于每个Q值单独跑一轮，一个Q值对应一个e+谱，从中随机抽样，依旧为中间Si(40um)发射，三块 Si大小及位置保持实验设置。
3. 每一组100000个事件。
4. 观察能谱时，由于中间Si太薄，考虑和前后其中一块Si符合。若不加条件，会出现两个峰。
5. 实验的四张谱分别表示：全谱，与后Si符合，与前SI符合，单谱。

```c++
%jsroot on

    TFile *f[7];
    int id[7]={5,7,8,9,10,11,13};
    TString fname[7]={"q5.root","q7.root","q8.root","q9.root","q10.root","q11.root","q13.root"};
    TH1D *h[7][4];
    char name[32];
    for(int i=0;i<7;i++) f[i]=new TFile("g4beta/"+fname[i]);
    TCanvas c;
    for(int i=0;i<7;i++){
        f[i]->cd();
        sprintf(name,"de[0]+de[1]+de[2]>>h%02d_111(1000,0,2)",id[i]);
        tree->Draw(name,"");
        sprintf(name,"h%02d_111",id[i]);
        h[i][0]=(TH1D*)f[i]->Get(name);
        sprintf(name,"Q=%02dMeV_111",id[i]);
        h[i][0]->SetNameTitle("beta energy spectrum 111",name);
        h[i][0]->SetLineColor(i*i+1);
        
        sprintf(name,"de[0]+de[1]+de[2]>>h%02d_011(1000,0,2)",id[i]);
        tree->Draw(name,"de[2]>0&&de[0]<0.01");
        sprintf(name,"h%02d_011",id[i]);
        h[i][1]=(TH1D*)f[i]->Get(name);
        sprintf(name,"Q=%02dMeV_011",id[i]);
        h[i][1]->SetNameTitle("beta energy spectrum 011",name);
        h[i][1]->SetLineColor(i*i+1);
        
        sprintf(name,"de[0]+de[1]+de[2]>>h%02d_110(1000,0,2)",id[i]);
        tree->Draw(name,"de[2]<0.01&&de[0]>0");
        sprintf(name,"h%02d_110",id[i]);
        h[i][2]=(TH1D*)f[i]->Get(name);
        sprintf(name,"Q=%02dMeV_110",id[i]);
        h[i][2]->SetNameTitle("beta energy spectrum 110",name);
        h[i][2]->SetLineColor(i*i+1);
        
        sprintf(name,"de[0]+de[1]+de[2]>>h%02d_010(1000,0,2)",id[i]);
        tree->Draw(name,"de[2]<0.01&&de[0]<0.01");
        sprintf(name,"h%02d_010",id[i]);
        h[i][3]=(TH1D*)f[i]->Get(name);
        sprintf(name,"Q=%02dMeV_010",id[i]);
        h[i][3]->SetNameTitle("beta energy spectrum 010",name);
        h[i][3]->SetLineColor(i*i+1);
    }

    h[6][0]->Draw();
    for(int i=0;i<6;i++) h[i][0]->Draw("same");
    c.BuildLegend();
    c.Draw();
```

```c++
c.Clear();
h[6][1]->Draw();
for(int i=0;i<6;i++) h[i][1]->Draw("same");
c.BuildLegend();
c.Draw();
```

```c++
c.Clear();
h[6][2]->Draw();
for(int i=0;i<6;i++) h[i][2]->Draw("same");
c.BuildLegend();
c.Draw();
```

```c++
c.Clear();
h[6][3]->Draw();
for(int i=0;i<6;i++) h[i][3]->Draw("same");
c.BuildLegend();
c.Draw();
```
